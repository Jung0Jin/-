```python
progresses = [93, 30, 55]
speeds = [1, 30, 5]
```

```python
import math
def solution(progresses, speeds):
    answer = []
    remain_day_list = []
    for i in range(len(progresses)):
        remain_progress = 100 - progresses[i]
        remain_day = remain_progress / speeds[i]
        remain_day_list.append(math.ceil(remain_day))
        
    for i in range(1, len(remain_day_list)):
        if remain_day_list[i-1] > remain_day_list[i]:
            remain_day_list[i] = remain_day_list[i-1]
            
    count = 1
    for i in range(1, len(remain_day_list)):
        if remain_day_list[i-1] >= remain_day_list[i]:
            count +=1
        else:
            answer.append(count)
            count = 1
    
    answer.append(count)
    return answer
```

```python
solution(progresses, speeds)
```

```python
def solution(progresses, speeds):
    Q = []
    for p,s in zip(progresses, speeds):
        if len(Q)==0 or Q[-1][0] < -((p-100)//s):
            Q.append([-((p-100)//s), 1])
        else:
            Q[-1][1] +=1
    return [q[1] for q in Q]
```

```python
solution(progresses, speeds)
```

```python
def solution(progresses, speeds):
    answer = []
    time = 0
    count = 0
    while len(progresses) > 0:
        if (progresses[0] + time * speeds[0]) >= 100:
            progresses.pop(0)
            speeds.pop(0)
            count += 1
        else:
            if count > 0:
                answer.append(count)
                count = 0
            time += 1
    answer.append(count)
    return answer
```

```python
solution(progresses, speeds)
```
