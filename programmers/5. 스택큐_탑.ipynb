```python
heights1 = [6,9,5,7,4]
heights2 = [3,9,9,3,5,7,2]
heights3 = [1,5,3,6,7,6,5]
```

```python
def solution(heights):
    answer = []
    for s_idx in range(len(heights)-1, -1, -1):
        flag = False
        for r_idx in range(s_idx-1, -1, -1):
            if heights[s_idx] < heights[r_idx]:
                flag = True
                answer = [r_idx + 1] + answer
                break
        if not flag:
            answer = [0] + answer
    return answer
```

```python
print(solution(heights1))
print(solution(heights2))
print(solution(heights3))
```

```python
def solution(heights):
    answer = [0] * len(heights)
    for i in range(len(heights)-1, 0, -1):
        for j in range(i-1, -1, -1):
            if heights[i] < heights[j]:
                answer[i] = j+1
                break
    return answer
```

```python
print(solution(heights1))
print(solution(heights2))
print(solution(heights3))
```

```python
def solution(heights):
    answer = [0] * len(heights)
    while heights:
        right = heights.pop()
        for idx in range(len(heights)-1, -1, -1):
            if heights[idx] > right:
                answer[len(heights)] = idx+1
                break
    return answer
```

```python
print(solution(heights1))
print(solution(heights2))
print(solution(heights3))
```
