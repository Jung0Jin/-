```python
clothes1 = [['yellow_hat', 'headgear'], ['blue_sunglasses', 'eyewear'], ['green_turban', 'headgear']]
clothes2 = [['crow_mask', 'face'], ['blue_sunglasses', 'face'], ['smoky_makeup', 'face']]
```

```python
def solution(clothes):
    spy = {} 
    answer = 1
    for i in clothes:
        if i[1] in spy:
            spy[i[1]].append(i[0])
        else:
            spy[i[1]] = [i[0]]
    for k in spy:
        answer *= len(spy[k]) + 1
    return answer - 1
```

```python
print(solution(clothes1))
print(solution(clothes2))
```

```python
from collections import Counter
from functools import reduce

def solution(clothes):
    cnt = Counter([kind for name, kind in clothes])
    answer = reduce(lambda x, y: x*(y+1), cnt.values(), 1) - 1
    return answer
```

```python
print(solution(clothes1))
print(solution(clothes2))
```

```python
def solution(clothes):
    d = dict()
    for _, body in clothes:
        d[body] = d.get(body, 0) + 1
    answer = 1
    for v in d.values():
        answer *= (v+1)
    return answer-1
```

```python
print(solution(clothes1))
print(solution(clothes2))
```

```python
from collections import Counter
def solution(clothes):
    cloth_list = []
    for cloth in clothes:
        cloth_list.append(cloth[1])
        answer = Counter(cloth_list)
    if len(answer) == 1:
        return list(answer.values())[0]
    else:
        ret = 1
        for v in list(answer.values()):
            ret *= (v+1)
        return ret-1
```

```python
print(solution(clothes1))
print(solution(clothes2))
```
